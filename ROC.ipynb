```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ROC.csv', delimiter=';')
```

```python
# TP - To są autoryzowane zalogowania uznane za poprawne 
# FN - to są autoryzowane zalogowania uznane za błędne zalogowania u nas FP więc zmieniam w csv na FN
# FP - To są nieautoryzowane zalogowania które powinny nie być uznane, ale są uznane za poprawne
# TN - To są ocbe zalogowania, uznane za nieporawne, czyli zakwalifikowane poprawnie
```

```python
df
```

```python
df['TPR'] = df['TP'] / (df['TP'] + df['FN'])
df['FPR'] = df['FP'] / (df['FP'] + df['TN'])
```

```python
df
```

```python
plt.figure()
plt.plot(df['FPR'], df['TPR'], color='darkorange', lw=2)
plt.plot([0,1], [0,1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('Receiver Operating Characteristic')
plt.show()
```
