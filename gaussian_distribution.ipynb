```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
df_result_authorized = pd.read_csv('punktacja_uprawniony.csv', delimiter=";")
df_result_unauthorized = pd.read_csv('punktacja_nieuprawniony.csv', delimiter=";")
```

```python
df_result_authorized['result'] = df_result_authorized['result'].str.replace(',', '.')
df_result_authorized['result'] = df_result_authorized['result'].astype(float)
df_result_unauthorized['result'] = df_result_unauthorized['result'].str.replace(',', '.')
df_result_unauthorized['result'] = df_result_unauthorized['result'].astype(float)
```

```python
df_result_unauthorized
```

```python
mean_authorized = df_result_authorized['result'].mean()
std_authorized = df_result_authorized['result'].std()

x_authorized = np.linspace(min(df_result_authorized['result']), max(df_result_authorized['result']), 100)
y_authorized = norm.pdf(x_authorized, mean_authorized, std_authorized)
```

```python
plt.figure()
plt.plot(x_authorized, y_authorized, label=f'result (śr: {mean_authorized:.2f}, std: {std_authorized:.2f})')
plt.title(f'Rozkład Gaussa dla punktacji uprawnionych')
plt.xlabel('Wyliczona wartość')
plt.ylabel('Gestość')
plt.legend()
plt.show()
```

```python
mean_unauthorized = df_result_unauthorized['result'].mean()
std_unauthorized = df_result_unauthorized['result'].std()

x_unauthorized = np.linspace(min(df_result_unauthorized['result']), max(df_result_unauthorized['result']), 100)
y_unauthorized = norm.pdf(x_unauthorized, mean_unauthorized, std_unauthorized)
plt.figure()
plt.plot(x_unauthorized, y_unauthorized, label=f'result (śr: {mean_unauthorized:.2f}, std: {std_unauthorized:.2f})')
plt.title(f'Rozkład Gaussa dla punktacji nieuprawnionych')
plt.xlabel('Wyliczona wartość')
plt.ylabel('Gestość')
plt.legend()
plt.show()
```
